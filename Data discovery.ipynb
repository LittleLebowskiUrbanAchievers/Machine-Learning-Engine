# Find numbers for each fight outcome

```python
f1Wins = len(fights[fights['f1result'] == 'win'])
fDraws = len(fights[fights['f1result'] == 'draw'])
fNC = len(fights[fights['f1result'] == 'NC'])
tfights = len(fights)

print("Wins:",f1Wins,'%' + '%.2f' % ((f1Wins / tfights) * 100))
print("Draws:",fDraws,'%' + '%.2f' % ((fDraws / tfights) * 100))
print("Not Complete:",fNC,'%' + '%.2f' % ((fNC / tfights) * 100))
print("Total Fights:",len(fights))
```
